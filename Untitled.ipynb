```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from ipwhois import IPWhois

lab4df = pd.read_csv('Lab4.csv', header = None, names = ['Type', 'sflow_agent_address', 'inputPort', 'outputPort',
               'src_MAC', 'dst_MAC', 'ethernet_type', 'in_vlan', 'out_vlan',
               'src_IP', 'dst_IP', 'IP_protocol', 'ip_tos', 'ip_ttl',
               'udp_src_port/tcp_src_port/icmp_type', 'udp_dst_port/tcp_dst_port/icmp_code', 'tcp_flags',
               'packet_size', 'IP_size', 'sampling_rate'])
lab4df.head()
```

```python
# Find the organisation of the IP Address
def organisation(ip):
    i = IPWhois(ip)
    ans = i.lookup_rdap()
    ans = ans.get("network", {}).get("name")
    return ans

def find_top(type):
    top = lab4df[type].value_counts().nlargest(5).to_frame()
    org = []
    for k in top.index:
        org.append(organisation(k))
    top["Organisation"] = org
    return top
```

```python
# Grouping by source and destination IP to find communication pairs
communication_pairs = lab4df.groupby(['src_IP', 'dst_IP']).size().reset_index(name='comm_count')

# Sorting by communication count in descending order to get top 5 pairs
top_5_pairs = communication_pairs.sort_values(by='comm_count', ascending=False).head(5)

# Displaying the top 5 communication pairs
print("Top 5 Communication Pairs:")
print(top_5_pairs)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Adding edges for each communication pair, with communication count as edge weight
for index, row in communication_pairs.iterrows():
    src, dst, count = row['src_IP'], row['dst_IP'], row['comm_count']
    G.add_edge(src, dst, weight=count)

# Create a layout for the graph
pos = nx.spring_layout(G, k=0.1, iterations=20)

# Draw the graph
plt.figure(figsize=(12, 12))
nx.draw(G, pos, with_labels=True, node_size=500, node_color='skyblue', font_size=10, font_weight='bold', alpha=0.7)
edge_weights = nx.get_edge_attributes(G, 'weight')

# Draw edge labels (communication count)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_weights, font_size=8)

plt.title("Network Communication between IP Hosts")
plt.show()
```
